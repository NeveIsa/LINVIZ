```python
def viz_rotate(theta):
    rotM = TFR.rotation(theta)
    VIZ.matrix(rotM,plot_std_bases=True)
    VIZ.show()
    
ICT.angle(viz_rotate)
```

```python
def viz_reflecion(theta):
    refM = TFR.reflection(theta)
    VIZ.matrix(refM,plot_std_bases=True)
    VIZ.show()
    
ICT.angle(viz_reflecion)
```

```python
def viz_projection(theta):
    projM = TFR.projection(theta)
    VIZ.matrix(projM,plot_std_bases=True)
    VIZ.show()
    
ICT.angle(viz_projection)
```
